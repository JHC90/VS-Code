# Erstelle Datensatz
Mit S-Klearn können synthethische Datenstätze selbst erzeugt werden. Das ist gut wenn man Algorithmen trainieren und verstehen möchte. Geht man her und erstellt die wahre Datenlage selbst, so kann später der Algorithmus gegen die Realität geprüft werden

## Cluster Datensatz

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 200, centers=4, cluster_std = 0.5, random_state = 0)
plt.scatter(X[:, 0], X[:, 1], s=50)
```
